```
from pyspark import SparkContext, SparkConf
conf = SparkConf().setAppName("Student Data Analysis")
sc = SparkContext.getOrCreate(conf = conf)
```

```
rdd = sc.textFile("/FileStore/tables/Student Data Files/StudentData.csv")
```

```
header = rdd.first()
rdd2 = rdd.filter(lambda x:x!=header)
```

```
rdd3  = rdd2.map(lambda x: x.split(','))
rdd4 = rdd3.map(lambda x : (x[1],int(x[5])))
total_marks = rdd4.reduceByKey(lambda x,y : x+y)
total_marks.collect()
```

```
rdd5 = rdd3.map(lambda x : int(x[5]))
passed = rdd5.filter(lambda x : x>=50)
passed.count()
```

```
failed = rdd5.filter(lambda x : x<50)
failed.count()
```

```
rdd6 = rdd3.map(lambda x : (x[3], (int(x[5]),1)))
rdd7 = rdd6.reduceByKey(lambda x,y:(x[0]+y[0] ,x[1]+y[1]))
number_of_students_per_course = rdd7.map(lambda x : (x[0], x[1][1]))
number_of_students_per_course.collect()
```

```
total_marks_by_course = rdd7.map(lambda x : (x[0], x[1][0]))
total_marks_by_course.collect()
```

```
avg_marks_by_course = rdd7.map(lambda x : (x[0], (x[1][0]/x[1][1])))
avg_marks_by_course.collect()
```

```
rdd8 = rdd3.map(lambda x : (x[3], x[5]))
min_marks_by_course = rdd8.reduceByKey(lambda x,y : x if x<y else y)
min_marks_by_course.collect()
```

```
max_marks_by_course = rdd8.reduceByKey(lambda x,y : x if x>y else y)
max_marks_by_course.collect()
```

```
rdd9 = rdd3.map(lambda x : (x[1],(int(x[0]),1)))
rdd10 = rdd9.reduceByKey(lambda x,y : ((x[0]+y[0]),(x[1]+y[1])))
avg_age_by_gender = rdd10.map(lambda x : (x[0] , x[1][0]/x[1][1]))
avg_age_by_gender.collect()
```
